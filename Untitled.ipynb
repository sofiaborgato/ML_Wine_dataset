```python
import sklearn
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np
from matplotlib.colors import ListedColormap
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.model_selection import ParameterGrid
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data=load_wine()
X=data["data"]
label=data["target"]
X = X[:,:2]
X.shape
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_C=scaler.fit_transform(X)




print(X_C)
```

```python
def plot_2d(X,label):
    cmap_bold = ListedColormap(['orange', 'teal', 'orangered'])
    fig,ax=plt.subplots(figsize=(6,6),dpi=90)
    ax.scatter(X[:,0],X[:,1],c=label,cmap=cmap_bold)
    
```

```python
plot_2d(X_C,label)
```

```python
X_train,X_tmp,Y_train,Y_tmp=train_test_split(X_C,label, test_size=0.5, random_state=18)
X_val,X_test,Y_val,Y_test=train_test_split(X_tmp,Y_tmp, train_size=0.4,random_state=19)
```

```python
accuracy_vect=[]
K=[1,3,5,7]

for k in K:
    classifier=KNeighborsClassifier(n_neighbors=k,weights='uniform')
    model=classifier.fit(X_train,Y_train)
    pred=classifier.predict(X_val)
    accuracy=(accuracy_score(pred,Y_val))
    accuracy_vect.append(accuracy)
    stringa="K="+str(k)
    plot_decision_boundaries(X_train,Y_train,model,stringa)

plt.figure()
plt.xlim([0,8])
plt.ylim([0,1])
plt.plot(K,accuracy_vect,color='blue',marker='o',linestyle='dashed')
plt.xticks(K)
plt.xlabel("K_neighbours")
plt.ylabel("Accuracy ")
plt.show()







    
    
    
```

```python
def plot_decision_boundaries(X_train,Y_train,model,stringa):

    h=0.07
    cmap_light = ListedColormap(['gold', 'mediumturquoise', 'salmon'])

    cmap_bold = ListedColormap(['orange', 'teal', 'orangered'])

    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(7,7))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X_train[:, 0], X_train[:, 1], c=Y_train, cmap=cmap_bold,
                edgecolor='k', s=70)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(stringa)

    plt.show()
    
    
    
```

```python
#TEST
K=5

classifier=KNeighborsClassifier(n_neighbors=K,weights='uniform')
classifier.fit(np.concatenate((X_train,X_val),axis=0),np.concatenate((Y_train,Y_val),axis=0))
pred=classifier.predict(X_test)
accuracy=accuracy_score(Y_test,pred)
print(accuracy)
```

```python
#K-FOLD
import matplotlib
X_train = np.concatenate((X_train, X_val))
Y_train = np.concatenate((Y_train, Y_val))
splits = 5
kf=KFold(n_splits=5)
C=[0.001,0.01,0.1,1,10,100,1000]
Gamma=[0.01,0.1,1,10,100]
params={"C":[0.001,0.01,0.1,1,10,100,1000],
        "gamma":[0.01,0.1,1,10,100],
       "kernel":['rbf']}
acc=[]
for config in ParameterGrid(params):
    svm_accuracy=[]
    count=[]
    for train_index, val_index in kf.split(X_train):
        xtrain, xtest = X_train[train_index], X_train[val_index]
        ytrain, ytest = Y_train[train_index], Y_train[val_index]
        count.append(len(train_index))
        model=SVC(**config)
        model.fit(xtrain,ytrain)
        accuracy=accuracy_score(ytest,model.predict(xtest))
        svm_accuracy.append(accuracy)
    acc.append(np.average(svm_accuracy,weights=count))
plot_grid_values(acc,C,Gamma)
acc = np.array(acc).reshape(len(C), len(Gamma))

plt.figure(figsize=(8, 6))
plt.subplots_adjust(left=.2, right=0.95, bottom=0.15, top=0.95)
plt.imshow(acc, interpolation='nearest', cmap='cubehelix_r')
plt.xlabel('Gamma')
plt.ylabel('C')
plt.colorbar()
plt.xticks(np.arange(len(Gamma)), Gamma, rotation=45)
plt.yticks(np.arange(len(C)), C)
plt.title('Validation accuracy')
plt.show()
acc  

           




```
